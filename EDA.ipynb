# EXPLORATORY DATA ANALYSIS

```python
import os
import pandas as pd
from skimage import io, color,transform
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
test_path = r'dataset\test'
train_path = r'dataset\train'
```

```python
cat_image =io.imread(test_path+r'\cats\cat_119.jpg')
dog_image =io.imread(test_path+r'\dogs\dog_130.jpg')
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(cat_image,cmap='gray')
plt.title("Sample Cat Image")

plt.subplot(1,2,2)
plt.imshow(dog_image,cmap='gray')
plt.title("Sample Dog Image")

plt.show()
```

```python
count_c = len(os.listdir(os.path.join(test_path, 'cats')) + os.listdir(os.path.join(test_path, 'cats')))
count_d = len(os.listdir(os.path.join(test_path, 'dogs')) + os.listdir(os.path.join(test_path, 'dogs')))
print("The number of cat images:", count_c)
print("The number of dog images:", count_d)
```

```python
sns.barplot(x=['Cats', 'Dogs'], y=[count_c, count_d])
plt.title('Distribution of Cats and Dog Images')
plt.show()
```

```python
cat_train_image_sizes = [io.imread(os.path.join('dataset/train/cats', img)).shape[0] for img in os.listdir('dataset/train/cats')]
dog_train_image_sizes = [io.imread(os.path.join('dataset/train/dogs', img)).shape[0] for img in os.listdir('dataset/train/dogs')]
cat_test_image_sizes = [io.imread(os.path.join('dataset/test/cats', img)).shape[0] for img in os.listdir('dataset/test/cats')]
dog_test_image_sizes = [io.imread(os.path.join('dataset/test/dogs', img)).shape[0] for img in os.listdir('dataset/test/dogs')]
```

```python
plt.figure(figsize=(18,8))
plt.title('Distribution of Image Heights for Training Dataset')

plt.subplot(1,2,1)
sns.histplot(data=cat_train_image_sizes, kde=True, label='Cats', color='blue' )
plt.xlabel('Image Height')
plt.ylabel('Frequency')

plt.subplot(1,2,2)
sns.histplot(data=dog_train_image_sizes, kde=True,  label='Dogs', color='orange')
plt.xlabel('Image Height')
plt.ylabel('Frequency')

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(18,8))
plt.title('Distribution of Image Heights for Testing Dataset')

plt.subplot(1,2,1)
sns.histplot(data=cat_test_image_sizes, kde=True, label='Cats', color='blue' )
plt.xlabel('Image Height')
plt.ylabel('Frequency')

plt.subplot(1,2,2)
sns.histplot(data=dog_test_image_sizes, kde=True,  label='Dogs', color='orange')
plt.xlabel('Image Height')
plt.ylabel('Frequency')

plt.legend()
plt.show()
```
